# Linear Regression Models - Simple Regression

```python
# Step 1 : import library
import pandas as pd
# Step 2 : import data
salary = pd.read_csv('https://github.com/prathvirao02/Datasets_used_for_AIML_Learning/raw/main/Salary%20Data.csv')
# Step 3 : define target (y) and features (X)
salary.columns
```

```python
y = salary['Salary']
X = salary[['Experience Years']]
# Step 4 : train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
y_pred
```

```python
# Step 8 : model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

# Mileage Prediction - Regression Analysis

###**Objective :**
To develop an AI/ML project using NumPy, pandas, Matplotlib, and Seaborn to predict vehicle mileage (MPG) through regression analysis. Key goals include data preprocessing, feature selection, model training, evaluation, and potentially deployment for accurate MPG estimation, benefiting the automotive industry.

###**Data Source :**
This dataset was taken from the StatLib library which is maintained at Carnegie Mellon University. The dataset was used in the 1983 American Statistical Association Exposition.

###**Import Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

###**Import Data**

```python
df = pd.read_csv('https://github.com/prathvirao02/Datasets_used_for_AIML_Learning/raw/main/MPG.csv')
df.head()
```

```python
df.nunique()
```

###**Data Preprocessing**

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

###**Remove Missing Values**

```python
df = df.dropna()
df.info()
```

###**Data Visualization**

```python
sns.pairplot(df, x_vars = ['displacement', 'horsepower', 'weight', 'acceleration', 'mpg'], y_vars = ['mpg']);
```

```python
sns.regplot(x = 'displacement', y = 'mpg', data = df);
```

###**Define Target Variable y and Feature X**

```python
df.columns
```

```python
y = df['mpg']
y.shape
```

```python
X = df[['displacement', 'horsepower', 'weight', 'acceleration']]
X.shape
```

```python
X
```

###**Scaling Data**

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X = ss.fit_transform(X)
X
```

```python
pd.DataFrame(X).describe()
```

**After Standardization Mean is Zero and Standard Deviation is One**

###**Train Test Split Data**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

###**Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

**Mileage = 23.4 - 1.05Displacement - 1.68Horsepower - 4.10Weight 0.115Acceleration + Error**

###**Predict Test Data**

```python
y_pred = lr.predict(X_test)
y_pred
```

###**Model Accuracy**

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

###**Polynomial Regression**

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_train2 = poly.fit_transform(X_train)
X_test2 = poly.fit_transform(X_test)
lr.fit(X_train2, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred_poly = lr.predict(X_test2)
```

###**Model Accuracy**

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
mean_absolute_error(y_test, y_pred_poly)
```

```python
mean_absolute_percentage_error(y_test, y_pred_poly)
```

```python
r2_score(y_test, y_pred_poly)
```
